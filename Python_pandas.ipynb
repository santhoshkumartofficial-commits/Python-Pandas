Pandas is a Python library used for data analysis, data cleaning, and data manipulation.




```python
import pandas

print("welcome pandas")
```

# 2. Creating DataFrames

From list of lists:

```python
df = pd.DataFrame([[1,2], [3,4]], columns=["A","B"])
```

**From dictionaries:**

```python
df = pd.DataFrame({
    "name": ["John", "Sara"],
    "marks": [85, 92]
})
```

**From CSV:**

```python
df = pd.read_csv("data.csv")
```

**To CSV:**

```python
df.to_csv("output.csv", index=False)
```

# 3. Basic DataFrame Operations

| Operation               | Syntax          |
| ----------------------- | --------------- |
| View first rows         | `df.head()`     |
| View last rows          | `df.tail()`     |
| Summary info            | `df.info()`     |
| Statistics              | `df.describe()` |
| Shape (#rows, #columns) | `df.shape`      |
| Column names            | `df.columns`    |


# 4. Selecting Data (Indexing)

**Column selection:**

```python
df["name"]
df[["name", "age"]]
```

Row selection:
**bold text**




```python
df.iloc[0]       # by index
df.loc[0]        # by label
```

**Slice rows:**

```python
df[1:4]
```

# 5. Filtering Data (Conditions)

```python
df[df["age"] > 30]
df[(df["marks"] >= 90) & (df["age"] < 25)]
df[df["city"].isin(["Delhi", "Chennai"])]
```

# 6. Adding Columns

```python
df["new_col"] = df["marks"] + 10
```

# 7. Deleting Columns/Rows
**Delete column:**

```python
df.drop("marks", axis=1, inplace=True)
```

**Delete row:**

```python
df.drop(2, axis=0)
```

# 8. Sorting

```python
df.sort_values("age")
df.sort_values(["age","marks"], ascending=[True, False])
```

# 9. Handling Missing Data
**Check missing:**

```python
df.isnull()
df.isnull().sum()
```

**Fill missing:**

```python
df.fillna(0)
df["age"].fillna(df["age"].mean())
```

**Drop missing:**

```python
df.dropna()
```

# 10. GroupBy (Very Important)

```python
df.groupby("city")["marks"].mean()
df.groupby("gender").agg({"marks": ["min","max","mean"]})
```

# 14. Value Counts
